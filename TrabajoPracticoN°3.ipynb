```python
################################################################################
############################OPERACION treePlot##################################
################################################################################

from graphviz import Digraph

class NodoArbolDeLibros(NodoArbolDeLibros):

  def treePlot(self, dot):
    if self.tieneIzquierdo():
      dot.node(str(self.izquierdo.autor), str(self.izquierdo.autor)+"\n"+str(self.izquierdo.libros))
      dot.edge(str(self.autor), str(self.izquierdo.autor))
      self.izquierdo.treePlot(dot)
    else:
      dot.node("None"+str(self.autor)+"l", "None")
      dot.edge(str(self.autor), "None"+str(self.autor)+"l")
    if self.tieneDerecho():
      dot.node(str(self.derecho.autor), str(self.derecho.autor)+"\n"+str(self.derecho.libros))
      dot.edge(str(self.autor), str(self.derecho.autor))
      self.derecho.treePlot(dot)
    else:
      dot.node("None"+str(self.autor)+"r", "None")
      dot.edge(str(self.autor), "None"+str(self.autor)+"r")

class ArbolDeLibros(ArbolDeLibros):
  
  def treePlot(self, fileName='tree'):
    if not self.estaVacio():
      treeDot = Digraph()
      treeDot.node(str(self.raiz.autor), str(self.raiz.autor+"\n"+str(self.raiz.libros)))
      self.raiz.treePlot(treeDot)
      treeDot.render(fileName, view=True)
```

```python

#TDA NodoArbol
class NodoArbolDeLibros:
  def __init__(self,nombreDeAutor):
    self.autor = nombreDeAutor
    self.izquierdo = None
    self.derecho = None
    self.libros = Lista() #LinkedList con los titulos de sus libros

  #se encarga de verificar si tiene un hijo izquierdo
  def tieneIzquierdo(self):
    return self.izquierdo != None

  #se encarga de verificar si tiene un hijo derecho
  def tieneDerecho(self):
    return self.derecho != None 

  #se encarga de saber el grado del nodo
  def grado(self):
    grado = 0
    if self.tieneIzquierdo():
      grado += 1 # si tiene un hijo izquierdo devuelve 1
    if self.tieneDerecho():
      grado += 1 # si tiene un hijo derecho devuelve 1
    return grado     

  #indica si no tiene hijos
  def esHoja(self):
    return self.grado() == 0  

  #muestra los datos del nodo primero el centro, luego el izquierdo y luego el derecho
  def preOrden(self):
    print(self.autor)
    if self.tieneIzquierdo():
      self.izquierdo.preOrden()
    if self.tieneDerecho():
      self.derecho.preOrden()
  #muestra los datos del nodo primero el izquierdo, luego el centro y luego el derecho
  def inOrden(self):
    if self.tieneIzquierdo():
      self.izquierdo.inOrden()
    print(self.autor)
    if self.tieneDerecho():
      self.derecho.inOrden()
  #muestra los datos del nodo primero el izquierdo, luego el centro y luego el derecho
  def postOrden(self):
    if self.tieneIzquierdo():
      self.izquierdo.postOrden()    
    if self.tieneDerecho():
      self.derecho.postOrden()
    print(self.autor)  

  #se encarga de insertar un nodo
  def insertar(self,nuevo):
    if nuevo.autor < self.autor: #si el dato es menor al dato central se inserta en el lado izquierdo
      if not self.tieneIzquierdo(): #si el lado izquierdo esta vacio de entrada, se inserta ahi
        self.izquierdo = nuevo
      else:
        self.izquierdo.insertar(nuevo) #si no esta vacio se llama recursivamente al lado izquierdo hasta encontrar uno vacio e insertarlo ahi
    elif nuevo.autor > self.autor: #si el dato es mayor al dato central se inserta en el lado derecho
      if not self.tieneDerecho(): #si el lado derecho esta vacio de entrada, se inserta ahi
        self.derecho = nuevo
      else:
        self.derecho.insertar(nuevo)  #si no esta vacio se llama recursivamente al lado derecho hasta encontrar uno vacio e insertarlo ahi
    else:
      raise Exception("El dato ya esta en el arbol") 

  def buscarAutor(self, dato):
    nodoDato = None 
    if self.autor == dato: #si el nodo actual tiene al autor se guarda el mismo nodo como dato
      nodoDato = self
    elif dato < self.autor: #si el dato es menor al dato del nodo actual
      if self.tieneIzquierdo(): #si tiene izquierdo entonces se busca del lado izquierdo al autor
        nodoDato = self.izquierdo.buscarAutor(dato)
    else:
      if self.tieneDerecho(): #si tiene derecho entonces se busca del lado derecho al autor
        nodoDato = self.derecho.buscarAutor(dato)
    return nodoDato


  def getNodo(self,autor):
    #Recibe un autor por parametro y devuelve el nodo en el que se encuentra
    nodoBuscado = None
    if self.autor == autor: #si el nodo actual tiene el autor, entonces se guarda el nodo actual
      nodoBuscado = self
    else:
      if autor < self.autor: #si el dato es menor lo busca por el lado izquierdo
        if self.tieneIzquierdo(): #si tiene nodo izquierdo
          nodoBuscado = self.izquierdo.buscarAutor(autor) #se llama recursivamente a la funcion del lado izquierdo
      else: #en el caso que no lo encuentre del lado izquierdo
        if self.tieneDerecho(): #si tiene derecho
          nodoBuscado = self.derecho.buscarAutor(autor) #se llama recursivamente al lado derecho a buscar al autor
    return nodoBuscado
  
  
  def maximo(self):
    nodoMax = None
    if self.tieneDerecho(): #en el caso q tenga derecho se va recursivamente al lado derecho
      nodoMax = self.derecho.maximo()
    else:
      nodoMax = self #se guarda a si mismo cuadno no tiene mas derechos
    return nodoMax 

  def minimo(self):
    nodoMin = None
    if self.tieneIzquierdo(): #en el caso q tenga izquierdo se va recursivamente al lado izquierdo
      nodoMin = self.izquierdo.minimo()
    else:
      nodoMin = self #se guarda a si mismo cuadno no tiene mas izquierdos
    return nodoMin

  def peso(self,peso=1):
    if self.tieneIzquierdo(): #si tiene izquierdo
      peso = 1 + self.izquierdo.peso() # se aumenta en 1 y se llama recursivamente hasta que no tenga mas izquierdos
    if self.tieneDerecho(): #si tiene derecho
      peso = 1 + self.derecho.peso() # se aumenta en 1 y se llama recursivamente hasta que no tenga mas derechos
    return peso

  def predecesor(self):
    prede = None
    if self.tieneIzquierdo():
      prede = self.izquierdo.maximo() #si tiene izquiedo va a ser su maximo
    return prede

  def sucesor(self):
    suces = None
    if self.tieneDerecho():
      suces = self.derecho.minimo() #si tiene derecho va a ser su minimo
    return suces
  
  #es la lista dentro del nodo
  def clonar(self,listaAClonar):
    if self.tieneIzquierdo(): #si tiene izquierdo
      listaAClonar.insertar(self.izquierdo.autor) #inserta en la lista clon el autpr del lado izquierdo
      self.izquierdo.clonar(listaAClonar) #se llama recursivamente hasta que no tenga mas izquierdo
    if self.tieneDerecho(): #si tiene derecho
      listaAClonar.insertar(self.derecho.autor)
      self.derecho.clonar(listaAClonar)

  def altura(self):
    alturaNodo = 0
    if self.grado() == 2: #si tiene 2 hijos
      alturaNodo = 1 + max(self.izquierdo.altura(),self.derecho.altura()) #guarda la suma de 1 y el maximo entre el nodo izquierdo y el nodo derecho
    elif self.tieneIzquierdo():
      alturaNodo = 1 + self.izquierdo.altura() # si tiene hijo izquiedo se llama recursivamente a la funcion
    elif self.tieneDerecho():
      altura = 1 + self.derecho.altura() # si tiene hijo derecho se llama recursivamente a la funcion
    return alturaNodo

  ###funciones para resolver objetivos###
  
  def insertarListaDeLibros(self,listaDeLibros):
    listaAux = listaDeLibros.clonar() #se clona la lista de entrada
    if self.libros.estaVacia(): #si los libros estan vacios
      while not listaAux.estaVacia(): #si no esta vacio 
        #self.libros.append(listaAux.obtenerElemento(0)) 
        #listaAux.eliminarElemento(0)
        self.libros.append(listaAux.pop(0)) #se agrega el elemento 0 de la lista 
    else:
      while not listaAux.estaVacia(): 
        if self.libros.contains(listaAux.obtenerElemento(0)):
        #if self.libros.buscaElemento(listaAux.obtenerElemento(0)) != [] :
          #self.libros.append(listaAux.obtenerElemento(0)) 
          #listaAux.eliminarElemento(0)
           self.libros.append(listaAux.pop(0)) #en el caso que contenga el elemento lo agrega y lo elimina de la lista aux
        else:
          #self.libros.append(listaAux.obtenerElemento(0) #si no )
          #listaAux.eliminarElemento(0)
          self.libros.append(listaAux.pop(0))
  
  

  def eliminarAutor(self,nombreAutor):
    if nombreAutor < self.autor: #en el caso de que el nombre del autor sea menor
      if self.tieneIzquierdo() and self.izquierdo.autor == nombreAutor:
        if self.izquierdo.grado() == 2: #si tiene 2 hijos
          nodoPredecesor = self.izquierdo.predecesor() #se guarda como predecesor el hijo izquierdo
          self.izquierdo.eliminarAutor(nodoPredecesor.autor) #se elimina el izquierdo
          nodoPredecesor.izquierdo = self.izquierdo.izquierdo #se asignan los hijos tanto izquierdo como derecho del predecesor
          nodoPredecesor.derecho = self.izquierdo.derecho
          self.izquierdo = nodoPredecesor #se asiga el nodo 
        elif self.izquierdo.tieneIzquierdo():
          self.izquierdo = self.izquierdo.izquierdo #se asigna el hijo izquierdo
        elif  self.izquierdo.tieneDerecho():
          self.izquierdo = self.izquierdo.derecho #se asigna el hijo derecho
        else:
          self.izquierdo = None
      elif self.tieneIzquierdo():
        self.izquierdo.eliminarAutor(nombreAutor) #se llama recursivamente a la funcion    
    elif  nombreAutor > self.autor: #en el caso que el nombre del autor sea mayor
      if self.tieneDerecho() and self.derecho.autor == nombreAutor:
        if self.derecho.grado() == 2: #si tiene 2 hijos
          nodoPredecesor = self.derecho.predecesor() #se guarda como predecesor el hijo derecho
          self.derecho.eliminarAutor(nodoPredecesor.autor) #se elimina el derecho
          nodoPredecesor.izquierdo = self.derecho.izquierdo #se asignan los hijos tanto izquierdo como derecho del predecesor
          nodoPredecesor.derecho = self.derecho.derecho
          self.derecho = nodoPredecesor  #se asiga el nodo 
        elif self.derecho.tieneIzquierdo():
          self.derecho = self.derecho.izquierdo #se asigna el hijo derecho
        elif self.derecho.tieneDerecho():
          self.derecho = self.derecho.derecho  #se asigna el hijo izquierdo
        else:
          self.derecho = None
      elif self.tieneDerecho():
        self.derecho.eliminarAutor(nombreAutor)  #se llama recursivamente a la funcion  


  def eliminarLibro(self,tituloLibro):
    if self.libros.contains(tituloLibro): #en el caso que la lista de libros contenga el titulo del libro
      posLibro = self.libros.posicion(tituloLibro) #se busca la posicion
      if posLibro != None:
        self.libros.eliminarElemento(posLibro) #se elimina el elemento en la posicion del libro
    if self.tieneIzquierdo():
      self.izquierdo.eliminarLibro(tituloLibro) #se tiene izquierdo se llama recursivamente para eliminar
    if self.tieneDerecho():
      self.derecho.eliminarLibro(tituloLibro) #se tiene derecho se llama recursivamente para eliminar
  

  def autoresConMasLibros(self,cantidadMinimaLibros):
    cont = 0
    if self.libros.len() >= cantidadMinimaLibros: #en el caso que sea mayor igual se da 1
      cont = 1
    if self.tieneIzquierdo():
      cont += self.izquierdo.autoresConMasLibros(cantidadMinimaLibros) #se llama recursivamente al hijo izquierdo
    if self.tieneDerecho():
      cont += self.derecho.autoresConMasLibros(cantidadMinimaLibros) #se llama recursivamente al hijo derecho
    return cont
  
  def cantidadTotalAutores(self,autor):
    cont  = 0
    if autor in self.autor: #si tiene el nombre del autor el contador es 1
      cont = 1              
    if self.tieneIzquierdo():
      cont = cont + self.izquierdo.cantidadTotalAutores(autor) #se llama recursivamente al lado izquierdo si es que tiene
    if self.tieneDerecho():
         cont = cont + self.derecho.cantidadTotalAutores(autor) #se llama recursivamente al lado derecho si es que tiene
    return cont     

  def librosEnNivel(self,nivel,listaDeLibros,cont = 0):
    if nivel == cont: #en el caso que sea igual
      aux = 0
      while aux < self.libros.len(): #mientras sea menor al largo de la lista entonces va agregnado a la lista de libros  el elemento obtener
        listaDeLibros.append(self.libros.obtenerElemento(aux))
        aux += 1
    if self.tieneIzquierdo() and nivel > cont:
      self.izquierdo.librosEnNivel(nivel,listaDeLibros,cont+1) # Se llama recursivamente al lado izquierdo si es que tiene y si nivel es mayor que cont, aumentando el cont + 1
    if self.tieneDerecho() and nivel > cont:
      self.derecho.librosEnNivel(nivel,listaDeLibros,cont+1)  # Se llama recursivamente al lado derecho si es que tiene y si nivel es mayor que cont, aumentando el cont + 1

  def internosAlfabetico(self,listaAutores):
    if self.tieneIzquierdo():   #se llama primero al hijo izquierdo porq su autor sera menor que el del nodo central
      self.izquierdo.internosAlfabetico(listaAutores) #llamada recursiva al hijo izquierdo
    if not self.esHoja():
      listaAutores.append(self.autor) #agrega a la lista el autor del nodo centro
    if self.tieneDerecho():
      self.derecho.internosAlfabetico(listaAutores) #se llama a la funcion en el hijo derecho    
  
  def autoresConPalabraEnTitulo(self, palabra, listaAutores):
    if self.libros.contienenPalabra(palabra): #pregunta si tiene la palabra
      listaAutores.append(self.autor) # si esta el autor se agrega el autor
    if self.tieneIzquierdo():
      self.izquierdo.autoresConPalabraEnTitulo(palabra, listaAutores) #se llama recursivamente al lado izquierdo si es que tiene
    if self.tieneDerecho():
      self.derecho.autoresConPalabraEnTitulo(palabra, listaAutores)#se llama recursivamente al lado derecho si es que tiene

##########################################################################################################################
#TDA ABB
class ArbolDeLibros:
  def __init__(self):
    self.raiz = None
  
  def estaVacio(self):
    return self.raiz == None #indica que est vacio si no tiene nada en la raiz

  def vaciar(self):
    self.raiz = None #vacia el arbol al poner en None la raiz

  def insertar(self, nuevo): 
    nodoNuevo = NodoArbolDeLibros(nuevo) #se genera un nodo nuevo con el dato
    if self.estaVacio():
      self.raiz = nodoNuevo #si la raiz esta vacia se inserta ahi
    else:
      self.raiz.insertar(nodoNuevo) #si no esta vacia la raiz se llama recursivamente hasta encontrar uno vacio 

  def preOrden(self):
    if not self.estaVacio():
      self.raiz.preOrden() #se llama a la funcion preOrden() del nodoArbol
  
  def inOrden(self):
    if not self.estaVacio():
      self.raiz.inOrden()#se llama la funcion inOrden() del nodoArbol
  
  def postOrden(self):
    if not self.estaVacio():
      self.raiz.postOrden() #se llama a la funcion postOrden() del nodoArbol

  def buscarAutor(self, dato): 
    estaEnArbol = False
    if not self.estaVacio():
      estaEnArbol = self.raiz.buscarAutor(dato) != None # se llama a la funcion buscarAutor del nodoArbol, y sera true si el dato es distinto de none, confirma que esta en el arbol
    return estaEnArbol

  def peso(self):
    cantNodos = 0
    if not self.estaVacio():
      cantNodos = self.raiz.peso() #se llama a la funcion de peso() del nodoArbol y se acumula el resultado
    return cantNodos

  def minimo(self):
    datoMinimo = None
    if not self.estaVacio():
      datoMinimo = self.raiz.minimo().dato #se llama a la funcion minimo() del nodoArbol y guarda el dato del nodo mas a la izquierda del arbol 
    return datoMinimo

  def maximo(self):
    datoMaximo = None
    if not self.estaVacio():
      datoMaximo = self.raiz.maximo().dato #se llama a la funcion maximo() del nodoArbol y guarda el dato del nodo mas a la derecho del arbol
    return datoMaximo

  def profundidad(self):
    prof = 0
    if not self.estaVacio():
      prof = self.raiz.altura() #se llama al afuncion altura() del nodoArbol y se guarda el valor
    return prof

  def treePlot(self, fileName='tree'):
    if not self.estaVacio():
      treeDot = Digraph()
      treeDot.node(str(self.raiz.dato), str(self.raiz.dato))
      self.raiz.treePlot(treeDot)
      treeDot.render(fileName, view=True)

  def getNodo(self, autor):
    nodoBuscado = None
    if not self.estaVacio():
      nodoBuscado = self.raiz.getNodo(autor) #se llama a la funcion getNodo() del nodoArbol para obtener el nodo buscado
    return nodoBuscado
  
  #Objetivos:
  #1
  def insertarLibros(self,listaLibros, nombreAutor):
    nuevoNodoDeAutor = NodoArbolDeLibros(nombreAutor) #se genera un nuevo nodo con el nombre del autor como dato
    if self.estaVacio():#si esta vacio 
      self.raiz = nuevoNodoDeAutor #se asigna a la raiz el nodo nuevo creado
      self.raiz.insertarListaDeLibros(listaLibros) #se inserta la lista de libros en el nodo del autor creado
    else:
      if self.buscarAutor(nombreAutor): #si no esta vacio se busca al autor y se lo inserta
        self.getNodo(nombreAutor).insertarListaDeLibros(listaLibros) 
      else:
        self.insertar(nombreAutor)#si no lo encuentra lo inserta
        self.getNodo(nombreAutor).insertarListaDeLibros(listaLibros)


  #2 
  def autoresConPalabraEnTitulo(self, palabra):
    listaAutores = Lista()
    if not self.estaVacio():
      self.raiz.autoresConPalabraEnTitulo(palabra, listaAutores) # llama a la funcion autoresConPalabraEnTitulo del nodoArbol pasando una lista que llena la funcion del nodoArbol
    return listaAutores


  #3 
  def buscarLibros(self,listaAutores):
     #Recibe una lista con los autores/as buscados/as. 
    #Retorna una lista con los títulos de los libros compartidos por todos/as los/as autores/as de la lista de entrada.
    #Si no hay ningun libro compatido por todos/as los/as autores/as, debe retornar
    #una lista vacía. Un libro compartido es un libro con el mismo título.
    lista = Lista()
    autoresNoEncontrados = 0
    aux = 0
    while aux < listaAutores.len(): #si el auxiliar es menor al largo de la lista de autores
      if not self.buscarAutor(listaAutores.obtenerElemento(aux)):
        autoresNoEncontrados += 1 
      aux += 1
    if autoresNoEncontrados == 0: #en el caso que esten todos los autores en el arbol
      aux = 1
      lista = self.getNodo(listaAutores.obtenerElemento(0)).libros #se guarda la lista del primer autor
      while aux < listaAutores.len():
        lista = lista.intersection(self.getNodo(listaAutores.obtenerElemento(aux)).libros) #se va recorriendo y se guarda la interseccion
        aux +=1 #se aumenta el auxiliar
    return lista
   
  #4
  def eliminarAutor(self,nombreAutor):
    if not self.estaVacio():
      if self.raiz.autor == nombreAutor:
        if self.raiz.grado() == 2: #en el caso que tengas los 2 hijos
          nodoPredecesor = self.raiz.predecesor()   #se guarda el predecesor 
          self.raiz.eliminarAutor(nodoPredecesor.autor) #se elimina el autor del nodoArbol
          nodoPredecesor.izquierdo = self.raiz.izquierdo # se guarda el valor del lado izquierdo
          nodoPredecesor.derecho = self.raiz.derecho #se guarda el valor del lado derecho
          self.raiz = nodoPredecesor  #se guarda el nodo en raiz 
        elif self.raiz.tieneIzquierdo(): #si tiene izquierdo se guarda el valor en la raiz
          self.raiz = self.raiz.izquierdo 
        elif self.raiz.tieneDerecho(): #si tiene derecho se guarda el valor en la raiz
          self.raiz = self.raiz.derecho
        else:
          self.raiz = None #si no hay nada queda none
      self.raiz.eliminarAutor(nombreAutor) #se llama a recursivamente
  
  
  #5
  def eliminarLibro(self,tituloLibro):
    if not self.estaVacio():
      self.raiz.eliminarLibro(tituloLibro) #si no esta vacio se llama a eliminar libro de nodoArbol
  

  #6
  def cantidadTotalAutores(self,nombre): #usar recorrido de busqueda con contador
    cont = 0
    if not self.estaVacio():
      cont = self.raiz.cantidadTotalAutores(nombre) #si no esta vacio se llama a cantidadTotalAutores de nodoArbol
    return cont
  
  #7
  def raizBalanceada(self):
    balance = None
    if not self.estaVacio():
      if self.raiz.grado() == 2: #si tiene ambos hijos
        balance = abs(self.raiz.izquierdo.peso() - self.raiz.derecho.peso()) <= 1  #se calcula que el peso izquierdo menos el derecho debe ser menor o igual a 1
      elif self.raiz.tieneIzquierdo():
        balance = self.raiz.izquierdo.peso() == 1 #si tiene un solo hijo es 1
      elif self.raiz.tieneDerecho():
        balance = self.raiz.derecho.peso() == 1 #si tiene un solo hijo es 1
    return balance
  
  #8
  def librosEnNivel(self,nivel):
    listaDeLibros = Lista()
    if not self.estaVacio():
      self.raiz.librosEnNivel(nivel,listaDeLibros) # llama a funcion de librosEnNivel del NodoArbol con la lista vacia creada
    return listaDeLibros
  #9
  def autoresConMasLibros(self,cantidadMinimaLibros):
    cont = 0
    if not self.estaVacio():
      cont = self.raiz.autoresConMasLibros(cantidadMinimaLibros) #se llama a la funcion autoresConMasLibros del NodoArbol
    return cont
  
  #10
  def internosAlfabetico(self):
    listaInterpretesAlfabeticos = Lista()
    if not self.estaVacio():
      self.raiz.internosAlfabetico(listaInterpretesAlfabeticos) #se llama a la funcion internosAlfabeticos de NodoArbol
    return listaInterpretesAlfabeticos
  
  

  
```

```python
#TDA Nodo
class NodoLista:
  def __init__(self, dato = None):
    self.dato = dato
    self.siguiente = None
  
  def tieneSiguiente(self):
    return self.siguiente != None

  def contains(self,dato):
    contains = False
    if self.dato == dato:
      contains = True
    elif self.tieneSiguiente():
      contains = self.siguiente.contains(dato)
    return contains  

  def posicion(self,dato,positActual=0):
    posit = None
    if dato == self.dato:
      posit = positActual
    else:
      posit = self.siguiente.posicion(dato,positActual+1)
    return posit

  def pop(self, posDel, posAct = 0):
    dato = None
    if posAct == posDel-1:
      dato = self.siguiente.dato
      self.siguiente = self.siguiente.siguiente
    else:
      dato = self.siguiente.pop(posDel, posAct+1)
    return dato

  def intersection(self,lista,intersection):
    if lista.contains(self.dato):
      if not intersection.contains(self.dato):
        intersection.append(self.dato)
    if self.tieneSiguiente():
      self.siguiente.intersection(lista,intersection)

#TDA Lista
class Lista:
  def __init__(self):
    self.primero = None
  
  def estaVacia(self):
    return self.primero == None

  def vaciar(self):
    self.primero = None

  def __repr__(self):
    strOut = "primero"
    actual = self.primero
    while actual != None:
      strOut += " -> " + str(actual.dato)
      actual = actual.siguiente
    strOut += " -|"
    return strOut
  
  def append(self, dato = None):
    nuevo = NodoLista(dato)
    if self.estaVacia():
      self.primero = nuevo
    else:
      actual = self.primero
      while actual.tieneSiguiente():
        actual = actual.siguiente
      actual.siguiente = nuevo

  def len(self):
    cont = 0
    actual = self.primero
    while actual != None:
      cont += 1
      actual = actual.siguiente
    return cont
    
  def pop(self, pos):
    if 0 <= pos < self.len():
      dato = None
      if pos == 0:
        dato = self.primero.dato  
        self.primero = self.primero.siguiente
      else:
        dato = self.primero.pop(pos)
      return dato
    else:
      raise Exception("Posicion incorrecta")
  
  def obtenerElemento(self, posDato):
    if 0 <= posDato < self.len():
      posActual = 0
      actual = self.primero
      while posActual < posDato:
        actual = actual.siguiente
        posActual += 1
      return actual.dato
    else:
      raise Exception("Posicion incorrecta")
    
  def insertarElemento(self, dato, posDato):
    nuevo = NodoLista(dato)
    if 0 < posDato < self.len():
      posActual = 0
      actual = self.primero
      while posActual < posDato-1:
        actual = actual.siguiente
        posActual += 1
      nuevo.siguiente = actual.siguiente
      actual.siguiente = nuevo
    elif posDato == 0:
      nuevo.siguiente = self.primero
      self.primero = nuevo
    elif posDato >= self.len():
      self.append(dato)
    else:
      raise Exception("Posicion incorrecta")

  def eliminarElemento(self, posDato):
    if 0 < posDato < self.len():
      posActual = 0
      actual = self.primero
      while posActual < posDato-1:
        actual = actual.siguiente
        posActual += 1
      actual.siguiente = actual.siguiente.siguiente
    elif not self.estaVacia() and posDato == 0:
      self.primero = self.primero.siguiente
    else:
      raise Exception("Posicion incorrecta")

  def clonar(self):
    nueva = Lista()
    actual = self.primero
    while actual != None:
      nueva.append(actual.dato)
      actual = actual.siguiente
    return nueva

  def intersection(self,lista):
    #Recibe una lista por parametro y devuelve la interseccion entre las 2 listas
    intersection = Lista()
    if not self.estaVacia():
      self.primero.intersection(lista,intersection)
    return intersection

  def contains(self,dato):
    #Recibe un dato por parametro y devuelve true si se encuentra en la lista
    contains = False
    if not self.estaVacia():
      contains = self.primero.contains(dato)
    return contains
  
  def posicion(self,dato):
    #Recibe un dato por parametro y devuelve la posicion en la que se encuentra en la lista.
    pos = None
    if not self.estaVacia() and self.contains(dato):
      pos = self.primero.posicion(dato)
    return pos


  #Reemplaza el dato de la lista por el dato pasado por parametro en la posicion dada. El viejo dato de la lista es removido permanentemente.
  def reemplazar(self,dato,posDato):
    self.eliminarElemento(posDato)
    self.insertarElemento(dato,posDato)

  def intercambiaPrimeros(self):
    primElem = self.primero.dato
    segElem = self.primero.siguiente.dato
    self.reemplazar(primElem,1)
    self.reemplazar(segElem,0)

  #Busca un elemento que recibe por parámetro. La operación retorna una lista con las ubicaciones del elemento en la lista de entrada.
  def buscaElemento(self,dato):
    posActual = 0
    finalList = []
    actual = self.primero
    while actual != None:
      if actual.dato == dato:
        finalList.append(posActual)
      posActual += 1
      actual = actual.siguiente
    return finalList
  
  def contienenPalabra(self, palabra):
    encontrada = False 
    aux = self.primero ## primer elemento de la lista de libros
    while aux != None: ## mientras no llegó al final de la lista
      if palabra in aux.dato: ## acá se usa la función 'in' de python que te permite saber si un string (palabra) está contenido dentro de otro string (aux.dato)
        encontrada = True
      aux = aux.siguiente
    return encontrada

```
